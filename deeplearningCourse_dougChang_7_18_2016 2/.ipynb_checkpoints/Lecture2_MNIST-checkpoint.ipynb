## Utility functions

```python
import pylab as pl
import cPickle, gzip
import numpy as np
from time import time
import matplotlib.pyplot as plt

def logistic(t):
    idx = t > 0
    out = np.copy(t)
    out[idx] = 1. / (1 + np.exp(-t[idx]))
    exp_t = np.exp(t[~idx])
    out[~idx] = exp_t / (1. + exp_t)
    return out

def dLogistic(x) :
    return x * (1-x)

def dTanh(x) :
    return 1 - np.power(x, 2)

def softmax(x):
    if (len(x.shape)==1):
        x=x.reshape(1,-1)
    x -= np.max(x,axis=1).reshape(-1,1) 
    expX=np.exp(x)
    return expX/np.sum(expX,axis=1).reshape((-1,1))

def add_bias(X):
    return np.hstack((np.ones((X.shape[0],1)),X))

def one_hot(y):
    if y.ndim == 1:
        y = y.reshape((-1,1))    
    return (y == np.unique(y)).astype(int)

def load_mnist_data(fname='mnist.pkl.gz'):
    
    train, valid, test = cPickle.load(gzip.open(fname,'rb'))
    
    X_train = add_bias(train[0])
    y_train = one_hot(train[1])
    
    X_test = add_bias(test[0])
    y_test = one_hot(test[1])
        
    return X_train, y_train, X_test, y_test

# plotting utilities
def get_figure(size=(12,8),xlabel=None,ylabel=None,title=None):
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1,1,1,axisbg = '#F8F8F8' )
    ax.grid(True)
    if xlabel is not None: ax.set_xlabel(xlabel)
    if ylabel is not None: ax.set_ylabel(ylabel)
    if title is not None: plt.title(title)     
    return ax

def plot_curve(x,y=None,size=(12,8),xlabel=None,ylabel=None,title=None,color='steelblue',alpha=0.8,marker=None): 
    ax = get_figure(size,xlabel,ylabel,title)
    if y is None:
        ax.plot(range(len(x)),x,color=color,alpha=alpha,marker=marker)
    else:
        ax.plot(x,y,color=color,alpha=alpha,marker=marker)
    plt.show()
```

# Batch GD (with fixed learning rate)

```python
activation = logistic       # hidden layer activation function
dActivation = dLogistic     # activation derivative for bprop

def predict_nnet(w1,w2,X):
    Z1   = activation(X.dot(w1))
    Z1   = add_bias(Z1)
    Z2   = Z1.dot(w2)     
    yHat = softmax(Z2)
    return (np.argmax(yHat,axis=1),yHat)

def loss_func(w1,w2,X,y,C):
    yHat = predict_nnet(w1,w2,X)[1]
    w    = np.hstack((w1.flat,w2.flat))
    loss = -np.sum(y * np.log(yHat))/X.shape[0] + 0.5*C*(w.dot(w))
    return loss

# load MNIST data
X_train, y_train, X_test, y_test = load_mnist_data()
N,D = X_train.shape
K   = len(y_train[0]) 

# tuning parameters
H           = 50     # number of first layer hidden neurons                                      
C           = 0      # L2 regularization constant
lrate       = 1      # learning rate a.k.a gradient step size
EPOCHS      = 200    # Max Number of training epochs
PRINT_EPOCH = 10     # printing loss every PRINT_EPOCH epochs

# init weights
w1 = np.random.uniform(-0.01,0.01,size=(D,H))   
w2 = np.random.uniform(-0.01,0.01,size=(H+1,K))

# accounting variables
train_loss_history     = []
train_accuracy_history = []
test_accuracy_history  = []

start_time = time()

print "================================================="
print "MNIST Character recognition data"
print "Number of training examples : ",N
print "Dimension of data(with bias): ",D
print "Number of distinct classes  : ",K 
print
print "Hidden Neurons  : ",H
print "L2 Reg          : ",C
print "Stepsize        : ",lrate
print "Epochs          : ",EPOCHS
print
print "Batch Gradient Descent"
print "================================================="
print

for epoch in range(EPOCHS):    
    X = X_train                      # Z0 = feature data
    y = y_train                      # label data in one hot encoding

    #         forward: fprop
    #==================================
    Z1   = activation(X.dot(w1))     # Z1
    Z1   = add_bias(Z1)              # Z1
    Z2   = Z1.dot(w2)                # Z2
    yHat = softmax(Z2)               # Z3 = yHat
    loss = loss_func(w1,w2,X,y,C)    # Z4 = loss replace with -np.mean(y * np.log(yHat))????

    #        backward: backprop
    #==================================
    delta2 = (yHat - y)/N
    dw2    = (Z1.T).dot(delta2) + C*w2
    
    delta1 = delta2.dot(w2[1:,:].T)
    delta1 = delta1 * dActivation(Z1[:,1:])
    dw1    = (X.T).dot(delta1) + C*w1  

    #         update weights 
    #==================================    
    w1 = w1 - lrate * dw1
    w2 = w2 - lrate * dw2 
    
    #    record loss/accuracy history
    #==================================
    train_accuracy = mean(np.argmax(yHat,axis=1)== np.argmax(y, axis=1))
    test_accuracy = mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))    
    train_loss_history.append(loss)
    train_accuracy_history.append(train_accuracy)
    test_accuracy_history.append(test_accuracy)

    if (epoch%PRINT_EPOCH == 0):
        print "Epoch ",epoch,", loss:",'{0:.4f}'.format(loss),\
               ", training accuracy:",'{0:.4f}'.format(train_accuracy),\
               ", testing accuracy:",'{0:.4f}'.format(test_accuracy)
print 
print "Final training accuracy: %0.4f"%mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
print "Final test accuracy    : %0.4f"%mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
print
elapsed = time() - start_time
print "%0.2f secs, " % (elapsed), '{0:.2f}'.format(elapsed/float(epoch+1)), "secs/epoch"

loss_history_Batch = train_loss_history
accuracy_history_Batch = train_accuracy_history
```

## momentum

```python
def dLogistic(x) :
    return x * (1-x)

activation = logistic       # hidden layer activation function
dActivation = dLogistic     # activation derivative for bprop

def predict_nnet(w1,w2,X):
    Z1   = activation(X.dot(w1))
    Z1   = add_bias(Z1)
    Z2   = Z1.dot(w2)     
    yHat = softmax(Z2)
    return (np.argmax(yHat,axis=1),yHat)

def loss_func(w1,w2,X,y,C):
    yHat = predict_nnet(w1,w2,X)[1]
    w    = np.hstack((w1.flat,w2.flat))
    loss = -np.sum(y * np.log(yHat))/X.shape[0] + 0.5*C*(w.dot(w))
    return loss

# load MNIST data
X_train, y_train, X_test, y_test = load_mnist_data()
N,D = X_train.shape
K   = len(y_train[0]) 

# tuning parameters
H           = 50     # number of first layer hidden neurons                                      
C           = 0      # L2 regularization constant
lrate       = 1      # learning rate a.k.a gradient step size
EPOCHS      = 200    # Max Number of training epochs
PRINT_EPOCH = 10     # printing loss every PRINT_EPOCH epochs

# init weights
w1 = np.random.uniform(-0.01,0.01,size=(D,H))   
w2 = np.random.uniform(-0.01,0.01,size=(H+1,K))

# accounting variables
train_loss_history     = []
train_accuracy_history = []
test_accuracy_history  = []


start_time = time()

print "================================================="
print "MNIST Character recognition data"
print "Number of training examples : ",N
print "Dimension of data(with bias): ",D
print "Number of distinct classes  : ",K 
print
print "Hidden Neurons  : ",H
print "L2 Reg          : ",C
print "Stepsize        : ",lrate
print "Epochs          : ",EPOCHS
print
print "Batch Gradient Descent"
print "================================================="
print

# momentum
gamma    = 0.9
dw1_prev = 0*w1
dw2_prev = 0*w2

for epoch in range(EPOCHS):    
    X = X_train                      # Z0 = feature data
    y = y_train                      # label data in one hot encoding

    #         forward: fprop
    #==================================
    Z1   = activation(X.dot(w1))     # Z1
    Z1   = add_bias(Z1)              # Z1
    Z2   = Z1.dot(w2)                # Z2
    yHat = softmax(Z2)               # Z3 = yHat
    loss = loss_func(w1,w2,X,y,C)    # Z4 = loss replace with -np.mean(y * np.log(yHat))????
    
    #        backward: backprop
    #==================================
    delta2 = (yHat - y)/N
    dw2    = (Z1.T).dot(delta2) + C*w2
    
    delta1 = delta2.dot(w2[1:,:].T)
    delta1 = delta1 * dActivation(Z1[:,1:])
    dw1    = (X.T).dot(delta1) + C*w1  

    #   update weights with momentum
    #==================================    
    w1 = w1 - lrate * dw1 - gamma * dw1_prev
    w2 = w2 - lrate * dw2 - gamma * dw2_prev
    
    dw1_prev = dw1
    dw2_prev = dw2

    #    record loss/accuracy history
    #==================================
    train_accuracy = mean(np.argmax(yHat,axis=1)== np.argmax(y, axis=1))
    test_accuracy = mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))    
    train_loss_history.append(loss)
    train_accuracy_history.append(train_accuracy)
    test_accuracy_history.append(test_accuracy)

    if (epoch%PRINT_EPOCH == 0):
        print "Epoch ",epoch,", loss:",'{0:.4f}'.format(loss),\
               ", training accuracy:",'{0:.4f}'.format(train_accuracy),\
               ", testing accuracy:",'{0:.4f}'.format(test_accuracy)
                
print 
print "Final training accuracy: %0.4f"%mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
print "Final test accuracy    : %0.4f"%mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
print
elapsed = time() - start_time
print "%0.2f secs, " % (elapsed), '{0:.2f}'.format(elapsed/float(epoch+1)), "secs/epoch"

loss_history_Batch_momentum     = train_loss_history
accuracy_history_Batch_momentum = train_accuracy_history
```

# SGD

```python
np.random.seed(0)

def dLogistic(x) :
    return x * (1-x)

activation = logistic       # hidden layer activation function
dActivation = dLogistic     # activation derivative for bprop

def predict_nnet(w1,w2,X):
    Z1   = activation(X.dot(w1))
    Z1   = add_bias(Z1)
    Z2   = Z1.dot(w2)     
    yHat = softmax(Z2)
    return (np.argmax(yHat,axis=1),yHat)

def loss_func(w1,w2,X,y,C):
    yHat = predict_nnet(w1,w2,X)[1]
    w    = np.hstack((w1.flat,w2.flat))
    loss = -np.sum(y * np.log(yHat))/X.shape[0] + 0.5*C*(w.dot(w))
    return loss

# load MNIST data
X_train, y_train, X_test, y_test = load_mnist_data()
N,D = X_train.shape
K   = len(y_train[0]) 

# tuning parameters
H          = 50     # number of first layer hidden neurons                                      
C          = 0      # L2 regularization constant
lrate      = 1      # learning rate a.k.a gradient step size
minibatch  = 64     # SGD mini-batch size
EPOCHS     = 10     # Max Number of training epochs

# init weights
w1 = np.random.uniform(-0.01,0.01,size=(D,H))   
w2 = np.random.uniform(-0.01,0.01,size=(H+1,K))

# accounting variables
train_loss_history_epoch = []
train_loss_history_SGD   = []
train_accuracy_history = []
test_accuracy_history  = []

start_time = time()

print "================================================="
print "MNIST Character recognition data"
print "Number of training examples : ",N
print "Dimension of data(with bias): ",D
print "Number of distinct classes  : ",K 
print
print "Hidden Neurons  : ",H
print "L2 Reg          : ",C
print "Stepsize        : ",lrate
print "Mini batch size : ",minibatch
print "Epochs          : ",EPOCHS
print
print "Stochastic Gradient Descent with Mini-batches"
print "================================================="
print

for epoch in range(EPOCHS):    
    for start, end in zip(range(0, N, minibatch), range(minibatch, N, minibatch)):
        X = X_train[start:end]     # Z0 = feature data
        y = y_train[start:end]     # label data in one hot encoding
    
        #         forward : fprop
        #==================================
        Z1   = activation(X.dot(w1))     # Z1
        Z1   = add_bias(Z1)              # Z1
        Z2   = Z1.dot(w2)                # Z2
        yHat = softmax(Z2)               # Z3 = yHat
        loss = loss_func(w1,w2,X,y,C)    # Z4 = loss
        train_loss_history_SGD.append(loss)
 
        #        backward: backprop
        #==================================
        delta2 = (yHat - y)/X.shape[0]
        dw2    = (Z1.T).dot(delta2) + C*w2
        
        delta1 = delta2.dot(w2[1:,:].T)
        delta1 = delta1 * dActivation(Z1[:,1:])
        dw1    = (X.T).dot(delta1) + C*w1  
    
        #         update weights 
        #==================================    
        w1 = w1 - lrate * dw1
        w2 = w2 - lrate * dw2 
        
    #    record loss/accuracy history
    #==================================
    train_accuracy = mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
    test_accuracy = mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
    loss = loss_func(w1,w2,X_train,y_train,C)    
    train_loss_history_epoch.append(loss)
    train_accuracy_history.append(train_accuracy)
    test_accuracy_history.append(test_accuracy)

    print "Epoch ",epoch,", loss:",'{0:.4f}'.format(loss),\
           ", training accuracy:",'{0:.4f}'.format(train_accuracy),\
           ", testing accuracy:",'{0:.4f}'.format(test_accuracy)
    
print 
print "Final training accuracy: %0.4f"%mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
print "Final test accuracy    : %0.4f"%mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
print
elapsed = time() - start_time
print "%0.2f secs, " % (elapsed), '{0:.2f}'.format(elapsed/float(epoch+1)), "secs/epoch"
```

# RMSProp

```python
def dLogistic(x) :
    return x * (1-x)

activation = logistic       # hidden layer activation function
dActivation = dLogistic     # activation derivative for bprop

def predict_nnet(w1,w2,X):
    Z1   = activation(X.dot(w1))
    Z1   = add_bias(Z1)
    Z2   = Z1.dot(w2)     
    yHat = softmax(Z2)
    return (np.argmax(yHat,axis=1),yHat)

def loss_func(w1,w2,X,y,C):
    yHat = predict_nnet(w1,w2,X)[1]
    w    = np.hstack((w1.flat,w2.flat))
    loss = -np.sum(y * np.log(yHat))/X.shape[0] + 0.5*C*(w.dot(w))
    return loss

def RMSprop(Eg_w,dw,rho,eps):
    Eg_w = rho*Eg_w + (1 - rho)*dw**2
    dw   = dw/np.sqrt(Eg_w + eps)
    return Eg_w,dw

# load MNIST data
X_train, y_train, X_test, y_test = load_mnist_data()
N,D = X_train.shape
K   = len(y_train[0]) 

# tuning parameters
H      = 50     # number of first layer hidden neurons                                      
C      = 0      # L2 regularization constant
lrate  = 0.001  # learning rate a.k.a gradient step size
minibatch  = 64     # SGD mini-batch size
EPOCHS = 10     # Max Number of training epochs
rho   = 0.9     # Gradient Sq weighting in RMSprop
eps   = 1e-6    # Epsilon to avoid div by zero

# init weights
w1 = np.random.uniform(-0.01,0.01,size=(D,H))   
w2 = np.random.uniform(-0.01,0.01,size=(H+1,K))

# accounting variables
train_loss_history_epoch = []
train_loss_history_SGD   = []
train_accuracy_history = []
test_accuracy_history  = []

start_time = time()

print "================================================="
print "MNIST Character recognition data"
print "Number of training examples : ",N
print "Dimension of data(with bias): ",D
print "Number of distinct classes  : ",K 
print
print "Hidden Neurons  : ",H
print "L2 Reg          : ",C
print "Stepsize        : ",lrate
print "Mini batch size : ",minibatch
print "Epochs          : ",EPOCHS
print
print "RMSprop adaptive lrate with rho: ",rho
print "================================================="
print

# RMSprop accumulators
Eg_w1 = 0*w1
Eg_w2 = 0*w2

for epoch in range(EPOCHS):    
    for start, end in zip(range(0, N, minibatch), range(minibatch, N, minibatch)):
        X = X_train[start:end]     # Z0 = feature data
        y = y_train[start:end]     # label data in one hot encoding
    
        #         forward : fprop
        #==================================
        Z1   = activation(X.dot(w1))     # Z1
        Z1   = add_bias(Z1)              # Z1
        Z2   = Z1.dot(w2)                # Z2
        yHat = softmax(Z2)               # Z3 = yHat
        loss = loss_func(w1,w2,X,y,C)    # Z4 = loss
        train_loss_history_SGD.append(loss)
 
        #        backward: backprop
        #==================================
        delta2 = (yHat - y)/minibatch
        dw2    = (Z1.T).dot(delta2) + C*w2
        
        delta1 = delta2.dot(w2[1:,:].T)
        delta1 = delta1 * dActivation(Z1[:,1:])
        dw1    = (X.T).dot(delta1) + C*w1  

        #             RMSprop
        #==================================            
        Eg_w1,dw1 = RMSprop(Eg_w1,dw1,rho,eps)
        Eg_w2,dw2 = RMSprop(Eg_w2,dw2,rho,eps)        
         
        #         update weights 
        #==================================    
        w1 = w1 - lrate * dw1
        w2 = w2 - lrate * dw2 
        
    #    record loss/accuracy history
    #==================================
    train_accuracy = mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
    test_accuracy = mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
    loss = loss_func(w1,w2,X_train,y_train,C)    
    train_loss_history_epoch.append(loss)
    train_accuracy_history.append(train_accuracy)
    test_accuracy_history.append(test_accuracy)

    print "Epoch ",epoch,", loss:",'{0:.4f}'.format(loss),\
           ", training accuracy:",'{0:.4f}'.format(train_accuracy),\
           ", testing accuracy:",'{0:.4f}'.format(test_accuracy)
        
print 
print "Final training accuracy: %0.4f"%mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
print "Final test accuracy    : %0.4f"%mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
print
elapsed = time() - start_time
print "%0.2f secs, " % (elapsed), '{0:.2f}'.format(elapsed/float(epoch+1)), "secs/epoch"
```

# Adadelta + ReLu + Dropout

```python
def Relu(x):
    return x*(x > 0)

def dRelu(x):
    return 1.0*(x > 0)

activation = Relu       # hidden layer activation function
dActivation = dRelu     # activation derivative for bprop

def dropout(X, p = 0.0):
    X *= np.random.binomial([np.ones(X.shape)],1-p)[0] * (1.0/(1-p))
    return X

def predict_nnet(w1,w2,X):
    Z1   = activation(X.dot(w1))
    Z1   = add_bias(Z1)
    Z2   = Z1.dot(w2)     
    yHat = softmax(Z2)
    return (np.argmax(yHat,axis=1),yHat)

def loss_func(w1,w2,X,y,C):
    yHat = predict_nnet(w1,w2,X)[1]
    w    = np.hstack((w1.flat,w2.flat))
    loss = -np.sum(y * np.log(yHat))/X.shape[0] + 0.5*C*(w.dot(w))
    return loss

def adadelta(Eg_w,Ed_w,dw,rho,eps):
    Eg_w = rho*Eg_w + (1 - rho)*dw**2
    dw   = np.sqrt(Ed_w + eps)*dw/np.sqrt(Eg_w + eps)
    Ed_w = rho*Ed_w + (1 - rho)*dw**2
    return Eg_w,Ed_w,dw

# load MNIST data
X_train, y_train, X_test, y_test = load_mnist_data()
N,D = X_train.shape
K   = len(y_train[0]) 

# tuning parameters
H      = 500    # number of first layer hidden neurons                                      
C      = 0      # L2 regularization constant
lrate  = 0.001  # learning rate a.k.a gradient step size
batch  = 64     # SGD mini-batch size
EPOCHS = 10     # Max Number of training epochs
rho   = 0.9     # Gradient Sq weighting in RMSprop
eps   = 1e-6    # Epsilon to avoid div by zero
p     = 0.5     # dropout probability

# init weights
w1 = np.random.uniform(-0.01,0.01,size=(D,H))   
w2 = np.random.uniform(-0.01,0.01,size=(H+1,K))

# accounting variables
train_loss_history_epoch = []
train_loss_history_SGD   = []
train_accuracy_history = []
test_accuracy_history  = []
start_time = time()

print "================================================="
print "MNIST Character recognition data"
print "Number of training examples : ",N
print "Dimension of data(with bias): ",D
print "Number of distinct classes  : ",K 
print
print "Hidden Neurons  : ",H
print "L2 Reg          : ",C
print "Mini batch size : ",batch
print "Epochs          : ",EPOCHS
print
print "Adadelta adaptive lrate with rho: ",rho
print "Hidden Layer with ReLu activation and dropout"
print "================================================="
print

# adadelta accumulators
Eg_w1 = 0*w1
Ed_w1 = 0*w1
Eg_w2 = 0*w2
Ed_w2 = 0*w2

for epoch in range(EPOCHS):    
    for start, end in zip(range(0, N, batch), range(batch, N, batch)):
        X = X_train[start:end]     # Z0 = feature data
        y = y_train[start:end]     # label data in one hot encoding
    
        #         forward : fprop
        #==================================
        Z1   = activation(X.dot(w1))     # Z1
        Z1   = add_bias(Z1)              # Z1
        Z1   = dropout(Z1,p)             # Z1 with "dropout"
        Z2   = Z1.dot(w2)                # Z2
        yHat = softmax(Z2)               # Z3 = yHat
        loss = loss_func(w1,w2,X,y,C)    # Z4 = loss
        train_loss_history_SGD.append(loss)
 
        #        backward: backprop
        #==================================
        delta2 = (yHat - y)/batch
        dw2    = (Z1.T).dot(delta2) + C*w2
        
        delta1 = delta2.dot(w2[1:,:].T)
        delta1 = delta1 * dActivation(Z1[:,1:])
        dw1    = (X.T).dot(delta1) + C*w1  

        #             adadelta
        #==================================            
        Eg_w1,Ed_w1,dw1 = adadelta(Eg_w1,Ed_w1,dw1,rho,eps)
        Eg_w2,Ed_w2,dw2 = adadelta(Eg_w2,Ed_w2,dw2,rho,eps)       
         
        #         update weights 
        #==================================    
        w1 = w1 - dw1
        w2 = w2 - dw2 
        
    #    record loss/accuracy history
    #==================================
    train_accuracy = mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
    test_accuracy = mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
    loss = loss_func(w1,w2,X_train,y_train,C)    
    train_loss_history_epoch.append(loss)
    train_accuracy_history.append(train_accuracy)
    test_accuracy_history.append(test_accuracy)

    print "Epoch ",epoch,", loss:",'{0:.4f}'.format(loss),\
           ", training accuracy:",'{0:.4f}'.format(train_accuracy),\
           ", testing accuracy:",'{0:.4f}'.format(test_accuracy)
        
print 
print "Final training accuracy: %0.4f"%mean(predict_nnet(w1,w2,X_train)[0]== np.argmax(y_train, axis=1))
print "Final test accuracy    : %0.4f"%mean(predict_nnet(w1,w2,X_test)[0]== np.argmax(y_test, axis=1))
print
elapsed = time() - start_time
print "%0.2f secs, " % (elapsed), '{0:.2f}'.format(elapsed/float(epoch+1)), "secs/epoch"
```
